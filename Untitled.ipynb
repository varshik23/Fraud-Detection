```python
def ip_add(ip):
    add=ip.split('.')
    return ''.join(add)
```

```python
IP_address['lower_bound_ip_address'] = IP_address['lower_bound_ip_address'].apply(ip_add)
IP_address['upper_bound_ip_address'] = IP_address['upper_bound_ip_address'].apply(ip_add)
```

```python
IP_address['lower_bound_ip_address'] = IP_address['lower_bound_ip_address'].map(int)
IP_address['upper_bound_ip_address'] = IP_address['upper_bound_ip_address'].map(int)
```

```python


# function that takes an IP address as argument and returns country associated based on IP_table

def IP_to_country(ip) :
    try :
        return IP_address.country[(IP_address.lower_bound_ip_address < ip)                            
                                & 
                                (IP_address.upper_bound_ip_address > ip)].iloc[0]
    except IndexError :
        return "Unknown"  
    
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
print('\n')
print('Accuracy:',accuracy_score(y_test,predictions))
```
